```python
# Restore the original 'runningScore' text data

# load the original file into a new, temporary DataFrame called 'temp_df'
print("Loading a temporary copy to restore the column...")
temp_df = pd.read_csv("Data/batting_card.csv")

# Copy ONLY the 'runningScore' column from the temporary copy 
# back into our main df_batting DataFrame. This replaces the column of NaNs
# with the original text data we need.
df_batting['runningScore'] = temp_df['runningScore']

print("'runningScore' column has been restored successfully!")

# VERIFY THE CHANGE
print("\nVerifying the change...")
df_batting.info()
```

```python
# This is a small helper function to convert the text to a dictionary.
# It handles any errors if the text is empty or badly formatted.
def safe_convert_to_dict(x):
    try:
        # This safely converts a string like "{'runs': 29}" into a real dictionary.
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return {}

#Create a temporary dictionary column
df_batting['score_dict'] = df_batting['runningScore'].apply(safe_convert_to_dict)


# Extract the 'runs' and 'wickets' numbers into new columns 
# .str.get('key') is the pandas way to get a value from a dictionary in a column.
df_batting['team_runs'] = df_batting['score_dict'].str.get('runs')
df_batting['wickets_down'] = df_batting['score_dict'].str.get('wickets')


#Clean up the new columns and delete the old ones

df_batting['team_runs'] = df_batting['team_runs'].fillna(0)
df_batting['wickets_down'] = df_batting['wickets_down'].fillna(0)

#Convert our new columns to integers (whole numbers).
df_batting['team_runs'] = df_batting['team_runs'].astype(int)
df_batting['wickets_down'] = df_batting['wickets_down'].astype(int)

#Delete the original and temporary columns.
df_batting.drop(columns=['runningScore', 'score_dict'], inplace=True)


# FINAL VERIFICATION
print("---------- Final Verification ----------")
print("Successfully created and cleaned 'team_runs' and 'wickets_down'.")
print("Old 'runningScore' column has been removed.")
df_batting.info()
```

```python
# For the 'commentary' column,replace missing values with an empty string,
# because no dismissal commentary exists for a 'not out' player.
print("Cleaning 'commentary' column...")
df_batting['commentary'] = df_batting['commentary'].fillna("")


# For 'runningOver',Will leave the missing values (NaN) as they are.
# This correctly represents that a 'not out' player was never dismissed in any over.
print("Leaving 'runningOver' missing values as NaN, as this is meaningful data.")
print("\n--- Final Check ---")
df_batting.isnull().sum()
```

```python
#Group the data by player's 'fullName' then select their 'runs' column and add them all with .sum().
top_scorers = df_batting.groupby('fullName')['runs'].sum()

#Their I sort the results from highest to lowest.
top_scorers = top_scorers.sort_values(ascending=False)

# Finally Select the first 10 rows from the sorted list.
print("--- Top 10 Run Scorers in the T20 World Cup of 2022---")
display(top_scorers.head(10))
```

```python
# Player Batting Summary: T20 World Cup 2022
print("Player Batting Summary: T20 World Cup 2022\n")

#First I want too add the Average column of Top 10 Run Scorers palyers.
#Average is calculated by how many time palyer is 'Out' or staying 'NotOut' at the end of match.
df_batting['was_out'] = ~df_batting['isNotOut']

#.agg() is used to doing the different calculations on different column at the same time.
player_summary = df_batting.groupby('fullName').agg(
    
    team_name = ('current_innings', 'first'),
    matches_played = ('match_id', 'nunique'),
    total_runs = ('runs', 'sum'),
    dismissals = ('was_out', 'sum')
)

#To calculate the average thats the formula.
player_summary['average'] = player_summary['total_runs'] / player_summary['dismissals']

# .loc() to find rows where average is infinity and update them.
#If Player was never out, average is just their total runs.
condition = player_summary['average'] == np.inf
player_summary.loc[condition, 'average'] = player_summary.loc[condition, 'total_runs']

display(player_summary.head())
```

```python
print("Now More Structured Top 10 Run Scorers of T20 WorldCup 2022\n")

# Sort the table by 'total_runs' from highest to lowest
final_table = player_summary.sort_values(by='total_runs', ascending=False)

# Select only the columns we want to show, in the order we want them.
final_table = final_table[['team_name', 'matches_played', 'average', 'total_runs']].round({'average': 2})

# Rename the columns to look Good for our final output
final_table.rename(columns={
    'team_name': 'Team',
    'matches_played': 'M',
    'average': 'Avg',
    'total_runs': 'Runs'
}, inplace=True)

# Set the Player's Name (which is the index) to be the main column
final_table.index.name = 'Player'

display(final_table.head(10))
```

```python
print("For find the Best Strike Rates Players.Add the column of ballsFaced column in our new \nplayer summary table.\n")
# Calculate total balls faced for each player

# group by player and sum up the 'ballsFaced' column
total_balls_faced = df_batting.groupby('fullName')['ballsFaced'].sum()

# add this new information as a column in our summary table
player_summary['total_balls_faced'] = total_balls_faced
display(player_summary.head())
```

```python
print("Top 10 Strike Rates of T20 WorldCup 2022 (min. 100 runs scored)\n")

# To calculate the overall tournament strike rate for each player
player_summary['overall_strike_rate'] = (player_summary['total_runs'] / player_summary['total_balls_faced']) * 100

# Filter who scored at least 100 runs
top_strikers = player_summary[player_summary['total_runs'] >= 100]

# Sort the filtered list by the new strike rate column, from highest to lowest
top_strikers = top_strikers.sort_values(by='overall_strike_rate', ascending=False)

# Select and rename columns for a good display table
final_strikers_table = top_strikers[['team_name', 'matches_played','total_runs',  'overall_strike_rate']].round(2)

final_strikers_table.rename(columns={
    'team_name': 'Team',
    'matches_played': 'M',
    'total_runs': 'Runs',
    'overall_strike_rate': 'Strike Rate'
}, inplace=True)

final_strikers_table.index.name = 'Player'
display(final_strikers_table.head(10))
```

```python
print("Top 10 Batting Averages of T20 WorldCup 2022 (min. 100 runs scored)\n")

# Filter who scored at least 100 runs
top_average_players = player_summary[player_summary['total_runs'] >= 100]

# Sort the filtered list by the 'average' column, from highest to lowest
top_average_players = top_average_players.sort_values(by='average', ascending=False)

# Select and rename columns for a good display table
final_average_table = top_average_players[['team_name', 'matches_played', 'total_runs', 'average']].round(2)

final_average_table.rename(columns={
    'team_name': 'Team',
    'matches_played': 'M',
    'total_runs': 'Runs',
    'average': 'Avg'
}, inplace=True)

final_average_table.index.name = 'Player'
display(final_average_table.head(10))
```

```python
print("Top 10 Highest Individual Scores of T20 WorldCup 2022 \n")

# let's first figure out who the opposition was for each innings.
df_batting['opposition'] = np.where(df_batting['current_innings'] == df_batting['home_team'], df_batting['away_team'], df_batting['home_team'])

# Sort the entire batting dataframe by the 'runs' column from highest to lowest.
highest_scores = df_batting.sort_values(by='runs', ascending=False)

# Select and rename columns for a good display table
final_table = highest_scores[['fullName', 'runs', 'strikeRate', 'current_innings', 'opposition', 'venue']].copy()

final_table = final_table.rename(columns={
    'fullName': 'Player',
    'runs': 'Score',
    'strikeRate': 'Strike Rate',
    'current_innings': 'Team',
    'opposition': 'Opposition',
    'venue': 'Venue'
})

top_10_scores = final_table.head(10)

# Set 'Player' as the index to make it bold and special.
top_10_scores = top_10_scores.set_index('Player')
display(top_10_scores)
```
